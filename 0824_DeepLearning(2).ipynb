```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
# 임의의 데이터 생성 -> data import

np.random.seed(0)  # 재현성을 위해 seed 설정
```

```python
# 100명의 환자 데이터 생성 (예: 나이, 종양 크기, 혈당 수치)

x = np.random.rand(100, 3)
```

```python
x
```

```python
# 임의로 생존 또는 사망 라벨 생성

y = np.random.randint(2, size=100)
```

```python
y
```

```python
# 간단한 딥러닝 모델 생성

model_example = Sequential()
```

```python
model_example.add(Dense(16, input_dim=3, activation='relu'))
```

```python
# 이진 분류를 위한 시그모이드 출력

model_example.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일 및 학습

model_example.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model_example.fit(x, y, epochs=100, verbose=0)
```

```python
# 모델 평가

loss, accuracy = model_example.evaluate(x, y)
```
